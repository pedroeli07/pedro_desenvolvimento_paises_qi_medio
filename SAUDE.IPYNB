```python
import pandas as pd
saude = pd.read_csv('HNP_StatsData.csv')
display(saude)
```

```python
paises_desejados = [
    'Estonia', 'Hungary', 'Poland', 'Lithuania', 'Croatia', 'Latvia', 'Italy',
    'Spain', 'Cyprus', 'Portugal', 'Malta', 'Mongolia', 'Bulgaria', 'Ukraine',
    'Moldova', 'Serbia', 'Iraq', 'Kazakhstan', 'Thailand', 'Armenia', 
    'Bosnia and Herzegovina', 'Costa Rica', 'Bhutan', 'Chile', 'Mexico', 
    'Tajikistan', 'Uruguay', 'Romania', 'Sri Lanka', 'Mauritius', 'Montenegro',
    'Georgia', 'Paraguay', 'Brazil', 'Colombia', 'Haiti', 'Dominican Republic',
    'Albania', 'Lebanon', 'Philippines', 'Peru', 'Jordan', 'Maldives', 'Pakistan',
    'Tunisia', 'Panama', 'Seychelles', 'Ecuador', 'Senegal', 'Bolivia', 'Uganda',
    'India', 'Algeria', 'Kenya', 'Angola', 'Tanzania', 'Bangladesh', 'Zimbabwe',
    'Burkina Faso', 'Mozambique', 'Burundi', 'Niger', 'Rwanda', 'Benin', 'Malawi',
    'El Salvador', 'South Africa', 'Lesotho', 'Zambia', 'Ethiopia', 'Nigeria',
    'Cameroon', 'Namibia', 'Central African Republic', 'Honduras', 'Mauritania',
    'South Sudan', 'Ghana', 'Nicaragua', 'Guatemala', 'Liberia', 'Sierra Leone',
    'Nepal'
]

# Filtre o DataFrame "saude" para manter apenas as linhas com países desejados
saude = saude[saude['Country Name'].isin(paises_desejados)]

# Exclua as colunas "Country Code" e "Indicator Code" do DataFrame "saude"
saude = saude.drop(columns=['Country Code', 'Indicator Code', 'Unnamed: 67'])

# Exiba as primeiras linhas do DataFrame "saude"
display(saude)
```

```python
variedade_indicator_names = saude['Indicator Name'].unique()

print(variedade_indicator_names)
```

```python
# Criar uma máscara booleana para as linhas que não contêm a palavra "married" no indicador
mask_married = ~saude['Indicator Name'].str.contains('married', case=False)

# Criar uma máscara booleana para as linhas que não contêm "Age population, age"
mask_age = ~saude['Indicator Name'].str.contains('Age population, age')

# Aplique a máscara para manter apenas as linhas desejadas
saude = saude[mask_married & mask_age]


# Agora, o DataFrame saude não contém as linhas dos indicadores de idade

variedade_indicator_names = saude['Indicator Name'].unique()

print(variedade_indicator_names)
```

```python
# Lista dos indicadores-chave de saúde em inglês
indicadores_saude = [
    'Maternal mortality ratio (modeled estimate, per 100,000 live births)',
    'Prevalence of HIV, total (% of population ages 15-49)',
    'Immunization, DPT (% of children ages 12-23 months)',
    'Immunization, measles (% of children ages 12-23 months)',
    'Births attended by skilled health staff (% of total)',
    'Health expenditure per capita (current US$)',
    'Health expenditure, public (% of government expenditure)',
    'Health expenditure, total (% of GDP)',
    'Mortality rate, under-5 (per 1,000 live births)',
]

# Filtrar o DataFrame mantendo apenas as linhas com indicadores de saúde relevantes
df_saude = saude[saude['Indicator Name'].isin(indicadores_saude)]
variedade_indicator_names = df_saude['Indicator Name'].unique()

print(variedade_indicator_names)
# Agora, df_saude contém apenas as linhas relacionadas aos indicadores de saúde.
display(df_saude)
```

```python
# Renomear as colunas do DataFrame
df_saude = df_saude.rename(columns={'Indicator Name': 'Indicador', 'Country Name': 'País'})

# Lista de indicadores de saúde relevantes em inglês
indicadores_saude_ingles = [
    'Maternal mortality ratio (modeled estimate, per 100,000 live births)',
    'Prevalence of HIV, total (% of population ages 15-49)',
    'Births attended by skilled health staff (% of total)',
    'Immunization, DPT (% of children ages 12-23 months)',
    'Immunization, measles (% of children ages 12-23 months)'
]

# Tradução dos nomes dos indicadores para o português
traducao_indicadores = {
    'Maternal mortality ratio (modeled estimate, per 100,000 live births)': 'Taxa de mortalidade materna',
    'Prevalence of HIV, total (% of population ages 15-49)': 'Prevalência de HIV/AIDS',
    'Births attended by skilled health staff (% of total)': 'Taxa de natalidade',
    'Immunization, DPT (% of children ages 12-23 months)': 'Cobertura de vacinação',
    'Immunization, measles (% of children ages 12-23 months)': 'Cobertura de vacinação de sarampo'
}

# Filtrar o DataFrame mantendo apenas as linhas com indicadores de saúde relevantes
df_saude = df_saude[df_saude['Indicador'].isin(indicadores_saude_ingles)]

# Traduzir os nomes dos indicadores para o português
df_saude['Indicador'] = df_saude['Indicador'].map(traducao_indicadores)

# Verificar a variedade de indicadores de saúde após a filtragem e tradução
variedade_indicator_names = df_saude['Indicador'].unique()
print(variedade_indicator_names)
```
