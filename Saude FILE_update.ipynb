```python
import pandas as pd
data = pd.read_csv("df_saude.csv")
print(data.columns)

data.head()
```

```python
# Seleciona as colunas desejadas
columns_to_select = ['País'] + [str(year) for year in range(1960, 2023)]
subset =data[columns_to_select]

# Encontra o último valor não nulo em cada linha
subset['Last_Value'] = subset.apply(lambda row: row.iloc[1:].last_valid_index(), axis=1)

# Cria um novo DataFrame com 'País' e 'Last_Value'
new_df = subset[['País', 'Last_Value']]
for index, row in new_df.iterrows():
    country_name = row['País']
    last_value = row['Last_Value']
    print(f"{country_name}: {last_value}")
```

```python
data.columns = data.columns.str.strip()
columns_to_exclude = [str(year) for year in range(1960, 2010)]

data = data.drop(columns=columns_to_exclude)

data.head()
```

```python
data.set_index(['País', 'Indicador'], inplace=True)

# Calcule a média das linhas (axis=1) excluindo a coluna do índice
data['Saude'] = data.mean(axis=1)

data.reset_index(inplace=True)
data.head()
```

```python
# Configure o Pandas para mostrar mais linhas
pd.set_option('display.max_rows', None)
print(data['Saude'])
```

```python
data.columns = data.columns.str.strip()
columns_to_exclude = [str(year) for year in range(2010, 2023)] + ['Unnamed: 0']
data = data.drop(columns=columns_to_exclude).round(2)
# Identifique os países que possuem algum valor NaN em 'Saude'
paises_com_nan = data[data['Saude'].isna()]['País'].unique()

# Remova os países com NaN em 'Saude'
data = data[~data['País'].isin(paises_com_nan)]
display(data)
```

```python
print(data.columns)
```

```python
import math
# Normalizar os indicadores de saúde para variar de 0 a 1
data['Saude_Normalizada'] = (data['Saude'] - data['Saude'].min()) / (data['Saude'].max() - data['Saude'].min())

# Defina os pesos para cada indicador
pesos = {
    'Taxa de natalidade': 1000,
    'Cobertura de vacinação': 2500,
    'Cobertura de vacinação de sarampo': 1000,
    'Taxa de mortalidade materna': -2750,
    'Prevalência de HIV/AIDS': -2750
}

# Calcular a pontuação de saúde ponderada e arredondar para cima
data['Pontuação Saúde'] = (
    data['Saude_Normalizada'] * data['Indicador'].map(pesos)
).apply(lambda x: math.ceil(x * 100) / 100)


# Exclua as colunas 'Indicador' e 'Saude'
data.drop(['Indicador', 'Saude_Normalizada'], axis=1, inplace=True)


# Mantenha apenas as colunas 'País' e 'Pontuação Saúde'
data = data[['País', 'Pontuação Saúde']]
# Exclua linhas repetidas mantendo apenas a primeira ocorrência
data.drop_duplicates(subset=['País'], inplace=True)

# Redefina o índice
data.reset_index(drop=True, inplace=True)
display(data)
```

```python
data.to_csv('df_saude_pontuacao.csv', index=False)
```
