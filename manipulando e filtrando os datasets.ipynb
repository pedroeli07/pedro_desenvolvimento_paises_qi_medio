```python
import pandas as pd
data = pd.read_csv("Saldo externo em bens e serviços (USD).csv")
print(data.columns)

data.head()
```

```python
# Seleciona as colunas desejadas
columns_to_select = ['Country Name'] + [str(year) for year in range(1960, 2023)]
subset = data[columns_to_select]

# Encontra o último valor não nulo em cada linha
subset['Last_Value'] = subset.apply(lambda row: row.iloc[1:].last_valid_index(), axis=1)

# Cria um novo DataFrame com 'Country Name' e 'Last_Value'
new_df = subset[['Country Name', 'Last_Value']]
for index, row in new_df.iterrows():
    country_name = row['Country Name']
    last_value = row['Last_Value']
    print(f"{country_name}: {last_value}")
```

```python
colunas_excluir = ['Country Code', 'Indicator Name', 'Indicator Code']
data = data.drop(colunas_excluir, axis=1)


data.columns = data.columns.str.strip()
columns_to_exclude = [str(year) for year in range(1960, 2010)] + ['Unnamed: 67']

data = data.drop(columns=columns_to_exclude)

data.head()
```

```python
# Defina a coluna "Country Name" como índice temporariamente
data.set_index('Country Name', inplace=True)

# Calcule a média das linhas (axis=1) excluindo a coluna do índice
data['Saldo externo em bens e serviços (USD)'] = data.mean(axis=1)

# Redefina o índice para o padrão (para que o "Country Name" não seja o índice)
data.reset_index(inplace=True)
data.head()
```

```python
# Configure o Pandas para mostrar mais linhas
pd.set_option('display.max_rows', None)
print(data['Saldo externo em bens e serviços (USD)'])
```

```python
data.columns = data.columns.str.strip()
columns_to_exclude = [str(year) for year in range(2010, 2023)]
data = data.drop(columns=columns_to_exclude)
data.head()
```

```python
data.dropna(subset=['Saldo externo em bens e serviços (USD)'], inplace=True)
```
